```python
import matplotlib.pyplot as plt
import numpy as np

# standard deviation of success rates and detection percentages
# Success rates as fractions for Raspberry Pi 4 for each attack type
rasp_pi_data_fractions = {
    'ARP Poisoning': [9/10, 8/10, 9/10, 10/10],  # Raspberry Pi 4 success rates for ARP Poisoning (as fractions)
    'DNS Cache Poisoning': [10/10, 8/10, 7/10, 9/10, 10/10],  # Raspberry Pi 4 success rates for DNS Cache Poisoning (as fractions)
    'LLMNR Poisoning': [10/10, 9/10, 10/10, 10/10],  # Raspberry Pi 4 success rates for LLMNR Poisoning (as fractions)
    'NTLM Relay Attack': [10/10, 9/10, 10/10, 10/10],  # Raspberry Pi 4 success rates for NTLM Relay Attack (as fractions)
    'IPv6 DNS Takeover': [10/10, 10/10, 10/10, 10/10]  # Raspberry Pi 4 success rates for IPv6 DNS Takeover (as fractions)
}

# Convert Raspberry Pi 4 success rates from fractions to percentages for each attack type
rasp_pi_data = {attack: [rate * 100 for rate in rates] for attack, rates in rasp_pi_data_fractions.items()}

# detection percentages by SOC tools for each attack type
detection_percentages = {
    'ARP Poisoning': [75, 80, 78, 72],  # Detection percentages for ARP Poisoning
    'DNS Cache Poisoning': [92, 85, 80, 88],  # Detection percentages for DNS Cache Poisoning
    'LLMNR Poisoning': [78, 82, 75, 80],  # Detection percentages for LLMNR Poisoning
    'NTLM Relay Attack': [60, 88, 85, 92],  # Detection percentages for NTLM Relay Attack
    'IPv6 DNS Takeover': [90, 88, 75, 92]  # Detection percentages for IPv6 DNS Takeover
}

# Calculate standard deviation for success rates and detection percentages
std_deviation_success = {attack: np.std(data) for attack, data in rasp_pi_data.items()}
std_deviation_detection = {attack: np.std(data) for attack, data in detection_percentages.items()}

# Create bar plots to visualize standard deviation for success rates and detection percentages
plt.figure(figsize=(12, 6))

# Plot for Success Rates
plt.subplot(1, 2, 1)
plt.bar(std_deviation_success.keys(), std_deviation_success.values(), color='skyblue')
plt.xlabel('Attack Type')
plt.ylabel('Standard Deviation')
plt.title('Standard Deviation of Success Rates for Raspberry Pi 4')
plt.xticks(rotation=45)

# Plot for Detection Percentages
plt.subplot(1, 2, 2)
plt.bar(std_deviation_detection.keys(), std_deviation_detection.values(), color='lightgreen')
plt.xlabel('Attack Type')
plt.ylabel('Standard Deviation')
plt.title('Standard Deviation of Detection Percentages by SOC Tools')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
